# Apriori Estimator
Implementation of Apriori Algorithm

```python
import pandas as pd
import numpy as np
import itertools
from sklearn.preprocessing import LabelEncoder
```

```python
data_path = '../data/store_data.csv'
```

```python
def flatten(array):
    """
    Returns a list o flatten elements of every inner lists (or tuples)
        ****RECURSIVE****
    """
    res = []
    for el in array:
        if isinstance(el, (list, tuple)):
            res.extend(flatten(el))
            continue
        res.append(el)
    return res
def is_subset_of(a, b):
    for da in a:
        if da not in b:
            return False
    return True
```

```python
class AprioriEstimator():
    
    def __init__(self, df_data):
        self.df_data = df_data
        self.dshape = df_data.shape
        self.le = LabelEncoder()
        self.extract_items(df_data)
    
    @property
    def data(self):
        return self.df_data
    @property
    def shape(self):
        return self.dshape
    
    @property
    def l1label(self):
        return zip(self.itemset, self.l1_count)
    
    def extract_items(self, data):
        self.raw_items = []
        for index, row in data.iterrows():
            row = pd.Series(row).dropna().values
            self.raw_items.append(row)
        self.itemset, self.l1_count = np.unique(flatten(self.raw_items), return_counts=True)
        self.itemset = pd.Series(self.itemset)
        self.le.fit(self.itemset)
        self.encoded_data = self.le.transform(self.raw_items)
        

    def apriori_gen(self, lk, n):
        c = flatten(lk)
        c = np.unique(c)
        return [(dc, 0) for dc in itertools.combinations(c, n) ]

    def ffis(self, min_sup):
        lk = self.le.transform(self.itemset)[self.l1_count > min_sup]
        i=2
        while len(lk) > 1:
            print(lk)
            ck = self.apriori_gen(lk, i)
            print(len(ck))
            for _, row in self.df_data.iterrows():
                for index, (c, c_coount) in enumerate(ck):
                    if is_subset_of(c, row.dropna()):
                        ck[index][1] += 1;
            lk = [ c  for c in ck if int(c[1]) > min_sup]
            i+=1
        return lk
            
        
        
        
```
